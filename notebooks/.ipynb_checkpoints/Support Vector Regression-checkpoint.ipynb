# Support Vector Regression

Support vector regression is a regression algorithm that supports (linear and non-linear). The main difference between SVM and SVR is that SVM is used for predicting discrete categorical labels while SVR is a regressor for predicting continuous ordered variables.

The idea in SVR is to approximate the best value within a given margin called $\epsilon-$ tube. Thus, the objective of SVR is to fit as many points as possible without violating that margin.

Keywords:

* **Hyperplane**: It is the line that helps in predeicting the target value. Is the separation between two data classes in a higher dimension.

* **Kernel**: Is the funciton that maps the data points into its higher dimension.

* **Boundary lines**: They are the thresholds. These lines are created in a distance $\epsilon$. It represents the margin between the data points.

* **Support Vector**: Extrem points in the datasets that defines the hyperlane.

By default:

```python
class sklearn.svm.SVR(*, kernel='rbf', degree=3, gamma='scale', coef0=0.0, tol=0.001, C=1.0, epsilon=0.1, shrinking=True, cache_size=200, verbose=False, max_iter=- 1)
```


`````python
from random import uniform,randint
import numpy as np
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

###############################################################################
# Generate datasets
def gen_coords(x1, y1, x2, y2, n):
    return [(uniform(x1, x2), uniform(y1, y2)) for _ in range(n)]

points = gen_coords(0,0,0.1,0.5,40)
x, y = zip(*points)

X = [ [value] for value in x]


###############################################################################
# Fit regression model
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
y_rbf = svr_rbf.fit(X, y).predict(X)

###############################################################################
# look at the results
import matplotlib.pyplot as plt
plt.scatter(X, y, c='k', label='data')
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, y_rbf -0.1, '--c',label=r'-e') # dashed cyan
plt.plot(X, y_rbf +0.1, '--c', label='+e') # dashed cyan
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()

score = svr_rbf.score(X,y)
print("R-squared:", score)
print("MSE:", mean_squared_error(y, y_rbf))
`````

As you can see in this example we generate a set of points using uniform distribution (*training set*). This way, **X** contains a list of lists which each element contains (*x-cordinate*) while **y** contains the (*y-cordinate*) for each point.


**Note**: The training set consists of the samples that were collected on which we have to perform prediction. We need to be sure that the features of the training set should suffice the domain that we are expecting.

## Kernel selection

Specifies the kernel type to be used in the algorithm. If none is given, ‘rbf’ will be used. If a callable is given it is used to precompute the kernel matrix.

```python
kernel{‘linear’, ‘poly’, ‘rbf’, ‘sigmoid’, ‘precomputed’} or callable, default=’rbf’
````

`````python
###############################################################################
# Generate sample data
import numpy as np

X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()

###############################################################################
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))
###############################################################################
# Fit regression model
from sklearn.svm import SVR

svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e4)
svr_poly = SVR(kernel='poly', C=1e4, degree=2)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)

###############################################################################
# look at the results
import matplotlib.pyplot as plt
plt.scatter(X, y, c='k', label='data')
plt.plot(X, y_rbf, c='g', label='RBF model')
plt.plot(X, y_lin, c='r', label='Linear model')
plt.plot(X, y_poly, c='b', label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()

score = svr_rbf.score(X,y)
print("(RBF) R-squared:" , score)
print("(RBF) MSE:", mean_squared_error(y, y_rbf))
score = svr_lin.score(X,y)
print("(LINEAR) R-squared:" , score)
print("(LINEAR) MSE:", mean_squared_error(y, y_lin))
score = svr_poly.score(X,y)
print("(POLY) R-squared:" , score)
print("(POLY) MSE:", mean_squared_error(y, y_poly))
`````

This example is extracted from the official documentation. It show how the different kernels affect SVR. Graphically, we can see that RBF is the better one; R-Squared corroborates it. RBF has an R-square value closer to 1.

Selection of kernel is important because if we choose a kernel like the Gaussian, which starts giving zero as the distance between the argument grows then as we start to move away from the training data, the machine will start returning the mean value of the training set. The selection of the kernel determines the estimator’s asymptotic behavior.
