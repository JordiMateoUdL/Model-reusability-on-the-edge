```python
home_dir = '../datasets'
dataset = 'GNFUV-USV-Dataset'
home_dir = home_dir +"/" + dataset
pi_dirs = os.listdir(home_dir)

data_list = []

columns = None


for pi_dir in pi_dirs:
    if 'pi' not in pi_dir:
        continue
    curr_dir = os.path.join(home_dir, pi_dir)
    data_file = os.path.join(curr_dir, os.listdir(curr_dir)[0])
    with open(data_file, 'r') as f:
        line = f.readline().strip().replace("'", '"')
        while line != '':
            try:
                input_json = json.loads(line)
                sensor_datetime = datetime.fromtimestamp(input_json['time'])
                input_json['time'] = sensor_datetime
                input_json['pi'] = pi_dir
                data_list.append(list(input_json.values()))
                if columns is None:
                    columns = list(input_json.keys())
            except JSONDecodeError as je:
                pass
            line = f.readline().strip().replace("'", '"')

            

data_df = pd.DataFrame(data_list, columns=columns)
del data_df['experiment']
del data_df['device']
del data_df['time']
del data_df['pi']
data_df = data_df.replace(to_replace=' None', value=np.nan).dropna()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

K=7
kmeans = KMeans(n_clusters=K).fit(data_df)
centroids = kmeans.cluster_centers_
print(centroids)
labels = ['Node{0}'.format(i) for i in range(K)]
plt_data = plt.scatter(data_df['humidity'], data_df['temperature'], c= kmeans.labels_.astype(float))
plt.colorbar()
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='x')
plt.title('K-means clustering')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
labels = ['Node{0}'.format(i) for i in range(K)]
for i in range (K):
    xy=(centroids[i, 0],centroids[i, 1])
    plt.annotate(labels[i],xy, horizontalalignment='right', verticalalignment='top',
          bbox=dict(boxstyle='round,pad=0.2', fc='yellow', alpha=0.3),
          arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.95', 
                            color='b'))
    
plt.show()
```

```python
samples = {kmeans.cluster_centers_[i, 0]: np.where(kmeans.labels_ == i)[0] for i in range(kmeans.n_clusters)}
```

```python
_samples={}
index = 0
for k in samples.keys():
    _samples[index]=samples[k]
    index = index + 1
```

```python
_samples
```

```python
# We build a dictionary for each cluster with the corresponding dataframe using the position obtained in Kmeans
nodes_dataset = {}
for k in _samples.keys():
    rows = data_df.iloc[_samples[k], :]
    nodes_dataset[k]=rows
nodes_dataset
```

```python
# We are going to train a model based on SVR
def SVR(df):
    # Fit regression model
    from sklearn.svm import SVR
    X=df.iloc[:,0:1].values
    y=df.iloc[:,1].values
    svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
    svr_rbf.fit(X, y)
    return svr_rbf

nodes_model={}
for k in nodes_dataset:
    nodes_model[k] = SVR(nodes_dataset[k])

# nodes_model contains the model trained with the own points of the node
nodes_model
```

```python
# We are going to generate a model with the own points + a % of the rest

# These are the outsiders, we are going to build a model with outer points for each cluster.
outer=14
nodes_dataset_with_outsiders={}
for k in nodes_dataset.keys():
    nodes_dataset_with_outsiders[k]=nodes_dataset[k].copy()
    for k1 in nodes_dataset.keys():
        if (k != k1):
            for o in range(outer):
                nodes_dataset_with_outsiders[k] =  pd.concat(
                    [nodes_dataset_with_outsiders[k], 
                     nodes_dataset[k1].sample()], ignore_index = True) 
            nodes_dataset[k1].reset_index()
nodes_dataset_with_outsiders
```

```python
nodes_model_with_outsiders={}
for k in nodes_dataset:
    nodes_model_with_outsiders[k] = SVR(nodes_dataset_with_outsiders[k])
nodes_model_with_outsiders
```

```python
from sklearn.metrics import mean_squared_error
#  cluster 2 -> 403      32.0         26.0
# El seu cluster és el 0
x = [[31]] 
#y_expected=[35]

### Model with all information
model = SVR(data_df)
y_expected = model.predict(x)
print(y_expected)

for k in nodes_model.keys():
    df = nodes_dataset[k]
    print("===== Node "+str(k))
    X=df.iloc[:,0:1].values
    y=df.iloc[:,1].values
    y_pred = nodes_model[k].predict(x)
    print(y_pred)
    score = nodes_model[k].score(X,y)
    print("R-squared:", score)
    print("MSE:", mean_squared_error(y_expected, y_pred))
    

for k in nodes_model.keys():
    df = nodes_dataset[k]
    print("===== Node with outsiders "+str(k))
    X=df.iloc[:,0:1].values
    y=df.iloc[:,1].values
    y_pred = nodes_model_with_outsiders[k].predict(x)
    print(y_pred)
    score = nodes_model_with_outsiders[k].score(X,y)
    print("R-squared:", score)
    print("MSE:", mean_squared_error(y_expected, y_pred))
```

```python
from sklearn.metrics import mean_squared_error
#  cluster 2 -> 403      32.0         26.0
# El seu cluster és el 0
x = [[31]] 

res={}
vals = data_df['humidity'].value_counts()
for v in vals.keys():
    x = [[v]]
    ### Model with all information
    model = SVR(data_df)
    y_expected = model.predict(x)
    
    res[v]={}
    res[v]["expected"]=y_expected
    node_res={}
    
    for k in nodes_model.keys():
        df = nodes_dataset[k]
        #print("===== Node "+str(k))
        X=df.iloc[:,0:1].values
        y=df.iloc[:,1].values
        y_pred = nodes_model[k].predict(x)
        #print(y_pred)
        score = nodes_model[k].score(X,y)
        #print("R-squared:", score)
        #print("MSE:", mean_squared_error(y_expected, y_pred))
        node_res[k]= mean_squared_error(y_expected, y_pred)

    res[v]["nodes"]=node_res
    node_res={}
    for k in nodes_model.keys():
        df = nodes_dataset[k]
        #print("===== Node with outsiders "+str(k))
        X=df.iloc[:,0:1].values
        y=df.iloc[:,1].values
        y_pred = nodes_model_with_outsiders[k].predict(x)
        #print(y_pred)
        score = nodes_model_with_outsiders[k].score(X,y)
        #print("R-squared:", score)
        #print("MSE:", mean_squared_error(y_expected, y_pred))
        node_res[k]= mean_squared_error(y_expected, y_pred)
    
    res[v]["nodes_with_outsiders"]=node_res
```
