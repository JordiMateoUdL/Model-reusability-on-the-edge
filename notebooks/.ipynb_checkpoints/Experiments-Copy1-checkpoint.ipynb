```python
home_dir = '../datasets'
dataset = 'GNFUV-USV-Dataset'
home_dir = home_dir +"/" + dataset
pi_dirs = os.listdir(home_dir)

data_list = []

columns = None


for pi_dir in pi_dirs:
    if 'pi' not in pi_dir:
        continue
    curr_dir = os.path.join(home_dir, pi_dir)
    data_file = os.path.join(curr_dir, os.listdir(curr_dir)[0])
    with open(data_file, 'r') as f:
        line = f.readline().strip().replace("'", '"')
        while line != '':
            try:
                input_json = json.loads(line)
                sensor_datetime = datetime.fromtimestamp(input_json['time'])
                input_json['time'] = sensor_datetime
                input_json['pi'] = pi_dir
                data_list.append(list(input_json.values()))
                if columns is None:
                    columns = list(input_json.keys())
            except JSONDecodeError as je:
                pass
            line = f.readline().strip().replace("'", '"')

            

data_df = pd.DataFrame(data_list, columns=columns)
del data_df['experiment']
del data_df['device']
del data_df['time']
del data_df['pi']
data_df = data_df.replace(to_replace=' None', value=np.nan).dropna()
```

```python
def generate_outsiders(input_Ds, strategy, outer, columns):
    
    """generate_outsiders
    Arguments:
        outer [scalar] -- number of points to replicate from the others.
        strategy [string] -- Values between ['random','percentile']
        input_Ds [ {id:[x_train,x_test,y_train,y_test]} ] -- Dictionary pointing each node with his data.
        columns [ [string] ] -- Array with the column names 
    Returns:
        {id:[x_train, y_train]} -- Dictionary pointing each node with his data + the outsiders.
    """
    
    output_Ds={}
    if (strategy == 'percentile'):
        percentile = ((100/outer)/100) - 0.01
        for k in input_Ds.keys():
            mk = np.array([input_Ds[k][0].ravel(),input_Ds[k][2]])
            dk = pd.DataFrame(mk.T,columns=columns)
            output_Ds[k]=dk.copy()
            for k1 in input_Ds.keys():
                if (k != k1):
                    mk1 = np.array([input_Ds[k1][0].ravel(),input_Ds[k1][2]])
                    dk1 = pd.DataFrame(mk1.T,columns=columns)
                    for o in range(outer):
                        output_Ds[k] =  pd.concat(
                            [output_Ds[k], 
                             pd.DataFrame(dk1.quantile( percentile + (percentile*o) )).T], ignore_index = True)
                        
    elif (strategy == 'random'):
        for k in input_Ds.keys():
            mk = np.array([input_Ds[k][0].ravel(),input_Ds[k][2]])
            dk = pd.DataFrame(mk.T,columns=columns)
            output_Ds[k]=dk.copy()
            for k1 in input_Ds.keys():
                if (k != k1):
                    mk1 = np.array([input_Ds[k1][0].ravel(),input_Ds[k1][2]])
                    dk1 = pd.DataFrame(mk1.T,columns=columns)
                    for o in range(outer):
                        output_Ds[k] =  pd.concat([output_Ds[k], pd.DataFrame(dk1.sample())], 
                                ignore_index = True)
                        
    else:
        raise ValueError("Wrong strategy. Strategy must be ['random','percentile']")
             
    return output_Ds

columns=['humidity','temperature']
#strategy='percentile'
input_Ds=nodes_dataset
outer=5

nodes_dataset_with_outsiders = generate_outsiders(input_Ds,'percentile',outer,columns)
nodes_dataset_with_outsiders_random = generate_outsiders(input_Ds,'random',outer,columns)
```

```python
# We are going to train a model based on SVR
def SVR(row): #row -> [x_train,x_test,y_train,y_test]
    # Fit regression model
    from sklearn.svm import SVR
    X=row[0].reshape(-1, 1) #x_train
    y=row[2] #y_train
    svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
    svr_rbf.fit(X, y)
    return svr_rbf

def SVR1(df):
    # Fit regression model
    from sklearn.svm import SVR
    X=(df.iloc[:,1].values).reshape(-1, 1)
    y=df.iloc[:,0].values
    svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
    svr_rbf.fit(X, y)
    return svr_rbf

def get_MSE(model,y_expected,x):
    from sklearn.metrics import mean_squared_error
    y_pred = model.predict(x.reshape(-1, 1))
    return mean_squared_error(y_expected, y_pred)
```

```python
#outersList = [5,10,15,20,25,30,35,40]
outersList = [5]
columns=['humidity','temperature']
input_Ds=nodes_dataset
K = 5 # This is the K value in the KMeans
    
# Full model represents the cloud with all the information available (full vision)
x=data_df.iloc[:,1].values
y=data_df.iloc[:,0].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
full_model = SVR([x_train,x_test,y_train,y_test])
    

experiments={}
for outer in outersList:
    # Step 1: Generate the outsiders using the outer value at each iteration
    nodes_dataset_with_outsiders = generate_outsiders(input_Ds,'percentile',outer,columns)
    nodes_dataset_with_outsiders_random = generate_outsiders(input_Ds,'random',outer,columns)
    
    # Step 2: Train the models
    nodes_model={}
    experiments[outer] = {}
    
    # For each test 
    for k in input_Ds.keys():
        
        experiments[outer][k] = {}
        
        # Valors per testejar el conjunt de test del node 0
        x = input_Ds[k][1] #x_test de node 0
        y_expected=input_Ds[k][3] #y_test de node 0
        
        # Train model without outsiders (only their data)
        for k1 in input_Ds.keys():
            nodes_model[k1] = SVR(input_Ds[k1])
        
        # Step 3: Make predictions
    
                
        # 3.1 using all data
        experiments[outer][k]['full']=get_MSE(full_model,y_expected,x)
        
        
        
        experiments[outer][k]['percentile']=[]
        experiments[outer][k]['random']=[]
        experiments[outer][k]['node']=[]
        nodes_with_outsiders_model={}
        nodes_with_outsiders_random={}
        for k1 in input_Ds.keys():
            #if k != k1:
            nodes_with_outsiders_model[k1] = SVR1(nodes_dataset_with_outsiders[k1])
            nodes_with_outsiders_random[k1] = SVR1(nodes_dataset_with_outsiders_random[k1])    
            
            # 3.3 using node data
            experiments[outer][k]['node'].append(get_MSE(nodes_model[k1],y_expected,x))
                
            # 3.1 percentile strategy
            experiments[outer][k]['percentile'].append(get_MSE(nodes_with_outsiders_model[k1],y_expected,x))
                
            # 3.2 random strategy
            experiments[outer][k]['random'].append(get_MSE(nodes_with_outsiders_random[k1],y_expected,x))

import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(experiments)
```

```python
# We are going to generate a model with the own points + a % of the rest

# These are the outsiders, we are going to build a model with outer points for each cluster.
outer=10

# Agant els percentils de la distribuci
percentile=0.09
nodes_dataset_with_outsiders={}
for k in nodes_dataset.keys():
    mk = np.array([nodes_dataset[k][0].ravel(),nodes_dataset[k][2]])
    dk = pd.DataFrame(mk.T,columns=['humidity','temperature'])
    nodes_dataset_with_outsiders[k]=dk.copy()
    for k1 in nodes_dataset.keys():
        if (k != k1):
            mk1 = np.array([nodes_dataset[k1][0].ravel(),nodes_dataset[k1][2]])
            dk1 = pd.DataFrame(mk1.T,columns=['humidity','temperature'])
            for o in range(outer):
                nodes_dataset_with_outsiders[k] =  pd.concat(
                        [nodes_dataset_with_outsiders[k], 
                         pd.DataFrame(dk1.quantile( percentile + (percentile*o) )).T], 
                    ignore_index = True) 
                
# Agafant mostres aleatories
nodes_dataset_with_outsiders_random={}
for k in nodes_dataset.keys():
    mk = np.array([nodes_dataset[k][0].ravel(),nodes_dataset[k][2]])
    dk = pd.DataFrame(mk.T,columns=['humidity','temperature'])
    nodes_dataset_with_outsiders_random[k]=dk.copy()
    for k1 in nodes_dataset.keys():
        if (k != k1):
            mk1 = np.array([nodes_dataset[k1][0].ravel(),nodes_dataset[k1][2]])
            dk1 = pd.DataFrame(mk1.T,columns=['humidity','temperature'])
            for o in range(outer):
                nodes_dataset_with_outsiders_random[k] =  pd.concat(
                        [nodes_dataset_with_outsiders_random[k], 
                         pd.DataFrame(dk1.sample())], 
                    ignore_index = True) 
    
print(nodes_dataset_with_outsiders)
```

```python
# We are going to train a model based on SVR
def SVR(row): #row -> [x_train,x_test,y_train,y_test]
    # Fit regression model
    from sklearn.svm import SVR
    X=row[0] #x_train
    y=row[2] #y_train
    svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
    svr_rbf.fit(X, y)
    return svr_rbf

def SVR1(df):
    # Fit regression model
    from sklearn.svm import SVR
    X=df.iloc[:,0:1].values
    y=df.iloc[:,0].values
    svr_rbf = SVR(kernel='rbf', C=1e4, gamma=0.1)
    svr_rbf.fit(X, y)
    return svr_rbf

# Train models without outsiders (only their data)
nodes_model={}
for k in nodes_dataset:
    nodes_model[k] = SVR(nodes_dataset[k])
    
# Train models with outsiders
nodes_with_outsiders_model={}
nodes_with_outsiders_random={}
for k in nodes_dataset:
    nodes_with_outsiders_model[k] = SVR1(nodes_dataset_with_outsiders[k])
    nodes_with_outsiders_random[k] = SVR1(nodes_dataset_with_outsiders_random[k])
    

# nodes_model contains the model trained with the own points of the node
#nodes_model
```
